```python
# Import modules
import sys
import numpy as np
import math
import re
import os
import subprocess
from collections import namedtuple
```

```python
# Functions
def create_name(root,date,case,nx,ny,nz,nn,nt,nr,nc):
    snx = str(int(nx)).zfill(4)
    sny = str(int(ny)).zfill(4)
    snz = str(int(nz)).zfill(4)
    #
    snn = str(int(nn)).zfill(4)
    snt = str(int(nt)).zfill(4)
    #
    out_name = root+"/"+date+"_"+case+snx+"x"+sny+"x"+snz+\
               "_NP"+snn+"x"+snt+\
               "_2D"+nr+"x"+nc
    return out_name
```

```python
def create_file_out(root,list_tests,date,nx,ny,nz,nr,nc,case):
    file_list=[]
    snx = str(int(nx)).zfill(4)
    sny = str(int(ny)).zfill(4)
    snz = str(int(nz)).zfill(4)
    ind = 0
    for test in list_tests:
        nome=root+"/"+date+'_Res'+snx+"x"+sny+"x"+snz+'_'+\
             test+'_D'+nr+'x'+nc+'_'+case+'.dat'
        ff=open(nome,'w')
        file_list.append(ff)
        if 'Tr' in test:
            str_write="# Nodes, Ptile,     TXY,      TYZ,     TZY,     TYX,     TOT\n"
        else:
            str_write="# Nodes, Ptile,    Forward,    Back\n"
        file_list[ind].write(str_write)
        ind = ind +1
    return file_list
```

```python
def close_file_out(file_list):
    for ii in range(len(file_list)):
        file_list[ii].close()
```

```python
def list_out(mypath):
    list_all=os.listdir(path=mypath)
    list_out=[]
    for nome in list_all:
        if 'listing' in nome:
            list_out.append(nome)
    return list_out
```

```python
def find_file(folder,file,list_files):
    file_out= folder+"/listing_"+file
    if not os.path.exists(file_out):
       file_out=[]
    return file_out
```

```python
def get_time_tr(open_file):
    time_out=np.zeros(5)
    ind = 0
    ref_file = open(open_file, 'r')
    for line in ref_file:
        if 'Avg Time ' in line:
            str_cut=line.split("   ")
            data = np.fromstring(str_cut[1],sep='  ',dtype='f8')
            time_out[ind] = data[0]
            ind = ind + 1
    ref_file.close()
    return time_out
```

```python
def get_time_fft(open_file):
    data=np.zeros(3)
    ref_file = open(open_file, 'r')
    for line in ref_file:
        if 'Avg time (sec):' in line:
            str_cut=line.split(":")
            dep = np.fromstring(str_cut[1],sep='  ',dtype='f8')
            data[0] = dep[0]
            data[1] = dep[1]
            data[2] = data[0]+data[1]
    ref_file.close()
    return data
```

```python
def write_line(nn,nt,time):
    str_write = str(nn)+', '+str(nt)
    for tt in time: 
        str_write = str_write+', '+str(tt)
    str_write = str_write+'\n'
    return str_write
```

```python
# Param definition
date='2025-10-22'
root='Leonardo_0512'
list_cases_in=['Resnvhpc_nccl',
               'Resnvhpc_nccl_even',
               'Resnvhpc_nccl_single',
               'Resnvhpc_nccl_even_single',
               'Resnvhpc_gpu',
               'Resnvhpc_gpu_even',
               'Resnvhpc_gpu_single',
               'Resnvhpc_gpu_even_single']
list_cases_out=['NCCL_ATOAV_D',               
                'NCCL_ATOA_D',
                'NCCL_ATOAV_S',
                'NCCL_ATOA_S',
                'CUMPI_ATOAV_D',               
                'CUMPI_ATOA_D',
                'CUMPI_ATOAV_S',
                'CUMPI_ATOA_S',]
nx=512
ny=512
nz=512
list_nn=[1, 1, 1, 2, 4]
list_nt = [1, 2, 4, 4, 4]
list_tests=['TrReal','TrClx','C2CX','R2CX','C2CZ','R2CZ']
list_tests_in=['timing2d_real','timing2d_complex','fft_c2c_x','fft_r2c_x','fft_r2c_z','fft_r2c_x']
#list_files=list_out('.')
#print(list_files)
```

```python
list_NR = ['0000','0001','NNNN']
list_NC = ['0000','NNNN','0001']
for kk in range(len(list_cases_in)):
    case_in = list_cases_in[kk]
    case_out = list_cases_out[kk]
    for jj in range(len(list_NR)):
        D1 = list_NR[jj]
        D2 = list_NC[jj]
        list_file_out= create_file_out(root,list_tests,date,nx,ny,nz,D1,D2,case_out)
        # Open the write file
        for ii in range(len(list_nn)):
            nn = list_nn[ii]
            nt = list_nt[ii]
            str_np = str(int(nn*nt)).zfill(4)
            if jj == 0 :
                nr = '0000'
                nc = '0000'
            elif jj == 1 :
                nr = '0001'
                nc = str_np
            else :
                nr = str_np
                nc = '0001'
            #
            folder = create_name(root,date,case_in,nx,ny,nz,nn,nt,nr,nc)
            list_files=list_out(folder)
            for ll in range(len(list_tests_in)):
                test = list_tests_in[ll]
                open_file = find_file(folder,test,list_tests_in)
                if len(open_file) > 0:
                    if 'fft' in test:
                        #Get time for FFT
                        time_fft = get_time_fft(open_file)
                        str_write = write_line(nn,nt,time_fft)
                    else:
                        time_tr = get_time_tr(open_file)
                        str_write = write_line(nn,nt,time_tr)
                    list_file_out[ll].write(str_write)         
        close_file_out(list_file_out)
```
